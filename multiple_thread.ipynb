```python
MODEL = "gpt-4o"  # You might want to make this configurable
client = OpenAI(api_key="")
```

```python
def call_llm_api(user_input,session_id):
    
    start_time=time.time()
    response=client.chat.completions.create(
            model=MODEL,
            messages=[
                {
                    "role": "system",
                    "content": "Answer the question appropriately"
                },
                {"role": "user", "content": user_input+session_id}
            ]
        )
    end_time=time.time()
    time_taken = end_time-start_time
    print(f"time taken for process:{time_taken}")
    
    return response.choices[0].message.content
```

```python
from azure.core.credentials import AzureKeyCredential
from azure.ai.documentintelligence import DocumentIntelligenceClient
from azure.ai.documentintelligence.models import AnalyzeResult, AnalyzeDocumentRequest

# For how to obtain the endpoint and key, please see PREREQUISITES above.
endpoint = "https://itranscenddiapi.cognitiveservices.azure.com/"
key = "58c11fea7f4345db8aea94e806e5ea22"

def get_di_result(formUrl,session_id):
    print(f"session_id:{session_id}")
    start_time=time.time()
    document_intelligence_client = DocumentIntelligenceClient(endpoint=endpoint, credential=AzureKeyCredential(key))

    poller = document_intelligence_client.begin_analyze_document(
        "prebuilt-layout",
        AnalyzeDocumentRequest(url_source=formUrl)
    )       


    result: AnalyzeResult = poller.result()
    end_time=time.time()
    time_taken=end_time- start_time
    print(f"time taken for process:{time_taken}")
    return result
```

```python
import sqlite3
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Step 1: Create and Populate Database
def setup_database():
    conn = sqlite3.connect("sales_data.db")
    cursor = conn.cursor()
    
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS SalesData (
        Date TEXT,
        Product TEXT,
        Sales INTEGER,
        Price REAL,
        MarketingSpend REAL,
        CompetitorPrice REAL,
        Seasonality INTEGER
    )""")
    
    sales_records = [
        ("2024-01-01", "A", 50, 10.5, 200, 9.8, 1),
        ("2024-01-02", "A", 55, 10.5, 250, 9.7, 1),
        ("2024-01-03", "A", 60, 10.0, 300, 9.9, 1),
        ("2024-01-04", "A", 58, 10.0, 280, 10.0, 1),
        ("2024-01-05", "A", 63, 9.8, 350, 10.2, 1)
    ]
    
    cursor.executemany("INSERT INTO SalesData VALUES (?, ?, ?, ?, ?, ?, ?)", sales_records)
    conn.commit()
    conn.close()

# Step 2: Function to generate SQL query
def generate_sql_query(product_name):
    return f"SELECT Date, Sales, Price, MarketingSpend, CompetitorPrice, Seasonality FROM SalesData WHERE Product = '{product_name}';"

# Step 3: Fetch data from database
def fetch_sales_data(product_name):
    conn = sqlite3.connect("sales_data.db")
    query = generate_sql_query(product_name)
    df = pd.read_sql_query(query, conn)
    conn.close()
    return df

# Step 4: Forecasting function
def predict_sales(df, future_days=7):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Days'] = (df['Date'] - df['Date'].min()).dt.days
    
    X = df[['Days', 'Price', 'MarketingSpend', 'CompetitorPrice', 'Seasonality']].values
    y = df['Sales'].values
    
    model = LinearRegression()
    model.fit(X, y)
    
    future_X = np.array([
        [df['Days'].max() + i, df['Price'].mean(), df['MarketingSpend'].mean(), df['CompetitorPrice'].mean(), df['Seasonality'].mode()[0]]
        for i in range(1, future_days + 1)
    ])
    predictions = model.predict(future_X)
    return predictions

# Step 5: Generate User-Friendly Response
def generate_response(product_name, predictions):
    avg_prediction = round(predictions.mean(), 2)
    return f"Based on previous trends, the estimated sales for {product_name} next week is around {avg_prediction} units."

# Step 6: End-to-End Execution
def main():
    setup_database()
    product_name = "A"
    sales_data = fetch_sales_data(product_name)
    predictions = predict_sales(sales_data)
    response = generate_response(product_name, predictions)
    print(response)

if __name__ == "__main__":
    main()
```

```python
import random
from datetime import datetime, timedelta
import pandas as pd
import numpy as np

def generate_supply_chain_data(start_date="2024-01-01", num_records=1000):
    # Set random seed for reproducibility
    random.seed(42)
    np.random.seed(42)
    
    # Initialize lists for each column
    dates = []
    products = []
    inventory_levels = []
    supplier_lead_times = []
    market_demands = []
    order_cancellations = []
    shipping_delays = []
    
    # Base values and trends
    base_inventory = 500
    base_demand = 200
    products_list = ["Product A", "Product B", "Product C"]
    
    # Generate data
    current_date = datetime.strptime(start_date, "%Y-%m-%d")
    
    for i in range(num_records):
        # Date
        dates.append(current_date.strftime("%Y-%m-%d"))
        
        # Product
        products.append(random.choice(products_list))
        
        # Inventory Level (with realistic fluctuations)
        inventory = max(0, base_inventory + np.random.normal(0, 50))
        inventory_levels.append(int(inventory))
        
        # Supplier Lead Time (typically 3-10 days with occasional spikes)
        lead_time = max(1, int(np.random.normal(6, 2)))
        supplier_lead_times.append(lead_time)
        
        # Market Demand (with upward trend and seasonal variation)
        trend = i * 0.1  # Slight upward trend
        seasonal = 20 * np.sin(2 * np.pi * i / 90)  # 90-day seasonality
        demand = max(0, base_demand + trend + seasonal + np.random.normal(0, 20))
        market_demands.append(int(demand))
        
        # Order Cancellations (correlated with demand)
        cancellation_rate = 0.05  # 5% of demand on average
        cancellations = max(0, int(demand * cancellation_rate * random.uniform(0.5, 1.5)))
        order_cancellations.append(cancellations)
        
        # Shipping Delays (1-5 days with occasional longer delays)
        delay = max(0, int(np.random.exponential(2)))
        shipping_delays.append(min(delay, 7))  # Cap at 7 days
        
        # Move to next date
        current_date += timedelta(days=1)
    
    # Create DataFrame
    df = pd.DataFrame({
        'Date': dates,
        'Product': products,
        'InventoryLevel': inventory_levels,
        'SupplierLeadTime': supplier_lead_times,
        'MarketDemand': market_demands,
        'OrderCancellations': order_cancellations,
        'ShippingDelays': shipping_delays
    })
    
    return df

# Generate the data
supply_chain_df = generate_supply_chain_data()

# Convert to list of tuples for SQL insertion
supply_records = list(supply_chain_df.itertuples(index=False, name=None))

# Print first few records as example
print("First 5 records:")
for record in supply_records[:5]:
    print(record)
```
